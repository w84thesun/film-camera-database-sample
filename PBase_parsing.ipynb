```python
import requests
from bs4 import BeautifulSoup
from bs4 import element
import csv
from lxml import etree
from pprint import pprint
import logging

base_url = "https://pbase.com"

cameras = []

logging.basicConfig(filename='bad_samples.log', encoding='utf-8', level=logging.INFO)

def get_brand_cameras(brand, cameras_link):
    cameras_page = requests.get(cameras_link)
    
    soup = BeautifulSoup(cameras_page.content, 'html.parser')

    cameras = soup.select('tr:nth-of-type(n+3)')
   
    for camera in cameras:
        try:
            camera_link = camera.contents[3].find('a')

            if type(camera_link) != element.Tag:
                print(camera)
                continue

            name = camera_link.text
            marketed = camera.contents[5].find('span').text

            r = {
            'brand': brand,
            'name':name,
            'marketed':marketed,
            'details': get_camera_details(base_url+camera_link.attrs['href']),
            }
        
        except:
            logging.info(camera)

        yield r        

def get_camera_details(camera_link):
    camera_page = requests.get(camera_link)
    
    soup = BeautifulSoup(camera_page.content, 'html.parser')
    
    details = soup.select('.small td')
    
    res = []
    
    for detail in details:
        text = detail.contents[0]
        
        if not text.isspace():
            res.append(text.strip())
            
    return ';'.join(res)    

page = requests.get(base_url+'/cameras')
soup = BeautifulSoup(page.content, 'html.parser')

brands = soup.select('td .ma a')

keys = ['brand', 'name', 'marketed', 'details']


with open('cameras.csv', 'w', newline='') as output_file:
    dict_writer = csv.DictWriter(output_file, keys)
    dict_writer.writeheader()

    for brand in brands:
        name = str(brand.contents[0].string)
        url = base_url+brand.attrs['href']
        
        for camera in get_brand_cameras(name, url):
            dict_writer.writerow(camera)
```
